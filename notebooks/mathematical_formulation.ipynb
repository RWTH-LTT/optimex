### Optimization Problem 

#### **Objective Function without forground/background differentiation:**

Minimize the environmental impact $h$, considering both the capacity-related and operation-related phases:

$$
\min_{S, O} h = \sum_{t \in T} \sum_{p \in P} \sum_{e \in E} Q_{e,t} \left[ \sum_{\tau \in \mathcal{T}, t - \tau \in T} B^{cap}_{p,e,\tau} S_{p,t - \tau} + B^{op}_{p,e} O_{p,t} \right]
$$

#### **Objective Function with forground/background differentiation:**

Minimize the direct environmental impact and indirect environmental impacts caused by the background system $h$, considering both the capacity-related and operation-related phases:
$$
\min_{S, O} h = \sum_{t \in T} \sum_{p \in P} \sum_{e \in E} Q_{e,t} \left[ \sum_{\tau \in \mathcal{T}, t - \tau \in T} B^{cap}_{p,e,\tau} S_{p,t - \tau} + B^{op}_{p,e} O_{p,t} \right. \\
+ \sum_{i \in I} \left( \sum_{\tau \in \mathcal{T}, t - \tau \in T} A^{cap}_{p,i,\tau} S_{p,t - \tau} + A^{op}_{p,i} O_{p,t} \right) \\
\left. \sum_{d \in D} M_{t,d} \left( B^{back}_{p,e,t} (A^{back}_{p,i,t})^{-1} \right) \right]
$$

---

#### **Constraints:**

1. **Operational Capacity:**

   $$
   O_{p,t} \leq \sum_{\tau \in \mathcal{T}, t - \tau \in T} A^{cap}_{p,r,\tau} S_{p,t - \tau}, \quad \forall p \in P, \; t \in T.
   $$

2. **Demand Satisfaction:**

   $$
   \sum_{p \in P} A^{op}_{p,r} O_{p,t} = F_{r,t}, \quad \forall r \in R, \; t \in T.
   $$

3. **Non-Negativity:**

   $$
   S_{p,t} \geq 0, \quad O_{p,t} \geq 0, \quad \forall p \in P, \; t \in T.
   $$

---

### **Indices:**

- **$p \in P$**: Processes.
- **$t \in T$**: Absolute time periods.
- **$\tau \in \mathcal{T}$**: Relative time periods .
- **$e \in E$**: Elementary flows.
- **$r \in R$**: Reference flows.
- **$i \in I$**: Intermediate flows.
- **$d \in D$**: Background databases.

---

### **Variables:**

- **$S_{p,t} \geq 0$**: Newly installed capacity for process $p$ at time $t$.
- **$O_{p,t} \geq 0$**: Operation level of process $p$ at time $t$.

---

### **Parameters:**

- **$Q_{e,t}$**: Environmental impact per unit of elementary flow $e$ at time $t$.
- **$B^{cap}_{p,e,\tau}$**: Environmental impact per unit of installed capacity for process $p$ and flow $e$ at time $\tau$.
- **$B^{op}_{p,e}$**: Environmental impact per unit of operation for process $p$ and flow $e$.
- **$A^{cap}_{p,r,\tau}$**: Output of reference flow $r$ per unit of installed capacity for process $p$ at time $\tau$.

  $$
  A^{cap}_{p,i,\tau} =
  \begin{cases}
    A_{p,i,\tau} & \text{if } \tau \in [\tau_{construction}, \tau_{construction} + \tau_{lifetime}] \\
    A_{p,i,\tau} & \text{if } i = r \\
    0 & \text{otherwise}
  \end{cases}
  $$

- **$A^{op}_{p,r}$**: Output of reference flow $r$ per unit of operation for process $p$.

  $$
  A^{op}_{p,i} = A^{op}_{p,i,\tau_{construction}}
  $$

- **$F_{r,t}$**: Demand for reference flow $r$ at time $t$.
- **$M_{t,d}$**: Mapping between databases $d$ and time $t$.
